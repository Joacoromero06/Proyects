```julia
function binomial_rv(n, p)
    successes = 0
    for i = 1:n
        if rand() < p
            successes += 1
        end
    end
    return successes
end
binomial_rv(5, 0.5)
```

Exercise 3 Compute an approximation to  using Monte Carlo
For random number generation use only rand()

```julia
function compute_pi(n)
    inside = 0
    for i = 1:n
        x, y = rand(), rand()
        if x ^ 2 + y ^ 2 <= 1 
            inside += 1
        end
    end
    # inside / n ----> area of circunference radius = 1, which is pi * 1/4
    return ( inside / n ) * 4
end
compute_pi(1000000000)
```

```julia
function flip()
    if rand()<0.5
        return true
    else 
        return false
    end
end
function three_heads_game(n)
    game = [flip() for i = 1:n]
    for i = 1:(n-2)
        if game[i] && game[i+1] && game[i+2]
            println("Sequence: $game")
            println("Winner")
            return 1
        end
    end
    println("Sequence: $game")
    println("Loose")
    return 0
end
three_heads_game(10)
```

```julia
function compute_population(filename)
    f = open(filename, "r")
    population = 0
    for line in eachline(f)
        splited_line = split(line, ":")
        population += parse(Int, splited_line[2])
    end
    return population
end
compute_population("cities_data.txt")
```

```julia
using LinearAlgebra
function discrete_lyapunov(A, sigma, epsilon=1e-5, max_iter=10000)
    Q = sigma * sigma'
    n = size(A,1)
    S = zeros(n,n)
    for i = 1:max_iter
        new_S = A * S * A' + Q
        if norm(new_S - S) < epsilon
            return new_S
        end
        S = new_S
    end
    error("no convergio")
end
A = [0.8 -0.2; -0.1 0.7]
sigma = [0.5 0.4; 0.4 0.6]
discrete_lyapunov(A,sigma) 
```

X0=x0 ,Xt+1 =aXt + b + t+1

```julia
using Distributions
using PyPlot
struct AR1{D<:Distribution}
    a::Real
    b::Real
    sigma::Real
    phi::D
end
function simulate_ts(m::AR1, x0:: Float64, n:: Int64)
    x = zeros(Float64, n)
    x[1] = x0

    for t = 1:n-1
        epsilon = rand(m.phi)
        x[t+1] = m.a * x[t] + m.b + m.sigma * epsilon
    end
    return x
end
m = AR1(0.98, 0.0, 1.2, Normal(0.0, 1.0))
ts = simulate_ts(m, 0.0, 100)
plot(ts, label="simulation")
legend()
xlabel("t")
ylabel("x_t")
title("time series")
```

Exercise 2 The term universal function is sometimes applied to functions which
 when called on a scalar return a scalar
 when called on an array of scalars return an array of the same length by acting elementwise
on the scalars in the array

For example, sin() has this property in Julia
julia> sin(pi)
1.2246467991473532e-16
julia> sin([pi, 2pi])
2-element Array{Float64,1}:
1.22465e-16
-2.44929e-16


Write a universal function f such that
f(k) returns a chi-squared random variable with k degrees of freedom when k is an integer
f(k_vec) returns a vector where f(k_vec)[i] is chi-squared with k_vec[i] degrees of free-
dom
Hint: If we take k independent standard normals, square them all and sum we get a chi-squared
with k degrees of freedom

```julia
function chi_squared(k::Int64)
    z = randn(k)
    return sum( z .^ 2 )
end
chi_squared(k_vec::AbstractVecOrMat{<:Int}) = chi_squared.(k_vec)
chi_squared([3,2,3])
```

```julia
u = Uniform()

n_rows, n_cols = 3, 2
fig, axes = subplots(n_rows, n_cols, figsize=(8,12))
subplot_num = 0

for i = 1:3
    for j = 1:2
        ax = axes[i,j]
        subplot_num += 1

        # generate normal sample with random mean and std
        m, s = 2rand(u) - 1, rand(u) + 1
        d = Normal(m, s)
        x = rand(d, 100)

        # histogram the sample
        ax[:hist](x, bins=20)
        ax[:set_title]("histogram $subplot_num")
        ax[:set_xticks]([-4, 0, 4])
        ax[:set_yticks]([])
    end
end
```

```julia
using QuantEcon: meshgrid

n = 50
x = range(-3, 3, n)
y = x
z = Array{Float64}(undef,n,n)
f(x, y) = cos(x^2 + y^2) / (1 + x^2 + y^2)
for i = 1:n
    for j = 1:n
        z[i,j] = f(x[i], y[j])
    end
end

fig = figure(figsize=(8,6))
ax = fig[:add_subplot](111, projection="3d")
ax[:set_zlim](-0.5, 1.0)
xgrid, ygrid = meshgrid(x, y)
ax[:plot_surface](xgrid, ygrid, z, cmap=ColorMap("jet"))
```

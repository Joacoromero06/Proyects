The binomial random variable Y  Bin(n, p) represents
 number of successes in n binary trials
 each trial succeeds with probability p

```julia
function binomial_rv(n, p)
    successes = 0
    for i = 1:n
        if rand() < p
            successes += 1
        end
    end
    return successes
end
binomial_rv(5, 0.5)
```

Exercise 3 Compute an approximation to  using Monte Carlo
For random number generation use only rand()

```julia
function compute_pi(n)
    inside = 0
    for i = 1:n
        x, y = rand(), rand()
        if x ^ 2 + y ^ 2 <= 1 
            inside += 1
        end
    end
    # inside / n ----> area of circunference radius = 1, which is pi * 1/4
    return ( inside / n ) * 4
end
compute_pi(1000000000)
```

 If U is a bivariate uniform random variable on the unit square (0, 1)2 , then the probability
that U lies in a subset B of (0, 1)2 is equal to the area of B
 If U1 , . . . , Un are iid copies of U, then, as n gets large, the fraction that falls in B converges to the probability of landing in B
 For a circle, area = pi * radius^2

```julia
function flip()
    if rand()<0.5
        return true
    else 
        return false
    end
end
function three_heads_game(n)
    game = [flip() for i = 1:n]
    for i = 1:(n-2)
        if game[i] && game[i+1] && game[i+2]
            println("Sequence: $game")
            println("Winner")
            return 1
        end
    end
    println("Sequence: $game")
    println("Loose")
    return 0
end
three_heads_game(10)
```

Exercise 5 Simulate and plot the correlated time series
x t +1 =  x t + e t +1

where
x0 = 0 and t = 0, . . . , T
The sequence of shocks {et} is assumed to be iid and standard normal
Set T = 200 and  = 0.9

```julia
using PyPlot
function simulation_ar1(x0, a, T)
    x = zeros(T+1)
    e = randn(T+1)
    for t = 0:T-1
        x[t+2] = a * x[t+1] + e[t+2]
    end
    return x
end

x0, alpha, T = 0, 1.5, 200
#plot(simulation_ar1(x0, alpha, T), label="AR(1)")
plot(simulation_ar1(x0, 1, T), label="AR(1) a = 1")
plot(simulation_ar1(x0, 0.98, T), label="AR(1) a = 0.98")
plot(simulation_ar1(x0, 0.8, T), label="AR(1) a = 0.8")
plot(simulation_ar1(x0, 0, T), label="AR(1) a = 0")

legend()
xlabel("t")
ylabel("x_t")
title("AR1: x_{t+1} =  x_{t} + e_{t}")
```

Exercise 4 Write a function that takes two sequences seq_a and seq_b as arguments and returns
true if every element in seq_a is also an element of seq_b, else false
 By sequence we mean an array, tuple or string

```julia
# is subset a of b ? 
function is_subset(a, b)
    for x in a
        if !(x in b)
            return false
        end
    end
    return true
end
is_subset([1,2,3], [1,2,3,4,5] )
```

Copy this text into a text file called us_cities.txt and save it in your present working directory

Write a program to calculate total population across these cities
Hints:
 If f is a file object then eachline(f) provides an iterable that steps you through the lines in
the file
 int("100") converts the string "100" into an integer

```julia
function compute_population(filename)
    f = open(filename, "r")
    population = 0
    for line in eachline(f)
        splited_line = split(line, ":")
        population += parse(Int, splited_line[2])
    end
    return population
end
compute_population("cities_data.txt")
```

Exercise 1 Consider the stochastic difference equation
Xt+1 = AXt + b + Wt     (1.2)

Here {Wt } is an iid vector of shocks with variance-covariance matrix equal to the identity matrix
Letting St denote the variance of Xt and using the rules for computing variances in matrix expres-
sions, it can be shown from (1.2) that {St } obeys
St+1 = ASt A0 + 0
(1.3)
Provided that all eigenvalues of A lie within the unit circle, the sequence {St } converges to a
unique limit S
This is the unconditional variance or asymptotic variance of the stochastic difference equation

This kind of equation is known as a discrete time Lyapunov equation

```julia
using LinearAlgebra
function discrete_lyapunov(A, sigma, epsilon=1e-5, max_iter=10000)
    Q = sigma * sigma'
    n = size(A,1)
    S = zeros(n,n)
    for i = 1:max_iter
        new_S = A * S * A' + Q
        if norm(new_S - S) < epsilon
            return new_S
        end
        S = new_S
    end
    error("no convergio")
end
A = [0.8 -0.2; -0.1 0.7]
sigma = [0.5 0.4; 0.4 0.6]
discrete_lyapunov(A,sigma) 
```

X0=x0 ,Xt+1 =aXt + b + t+1

```julia
using Distributions
using PyPlot
struct AR1
    a::Real
    b::Real
    sigma::Real
    phi::Distribution
end
function simulate_ts(m::AR1, x0:: Float64, n:: Int64)
    x = zeros(Float64, n)
    x[1] = x0

    for t = 1:n-1
        epsilon = rand(m.phi)
        x[t+1] = m.a * x[t] + m.b + m.sigma * epsilon
    end
    return x
end
m = AR1(0.98, 0.0, 1.2, Laplace())
ts = simulate_ts(m, 0.0, 1000)
plot(ts)
```

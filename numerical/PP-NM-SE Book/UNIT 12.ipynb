```python
x1 = np.linspace( -3, 3, 100 )
x2 = np.linspace( -6, 6, 200 )
plt.style.use('seaborn-v0_8-poster')
fig, axes = plt.subplots( nrows = 2, ncols = 1, figsize = ( 6, 9 ) )
ax1, ax2 = axes

# PLOTTING DATA
ax1.plot( x1, np.exp(x1), 'ro', label = 'exponential' )
ax2.plot( x2, x2**2 + 2*x2 + 4, 'm.', label = 'polinomical' )

# MAKING BEATIFUL
ax1.set_xlabel('X axis', fontsize = 15)
ax1.set_ylabel('Y axis', fontsize = 15)
ax1.set_title('Plot of:  e^x function  ', fontsize = 13)
ax1.legend(loc = 'upper left', title = 'Info of functions')
ax1.set_xlim(-3, 3)
ax1.set_ylim( min(np.exp(x1)), max(np.exp(x1)))
ax1.grid()

ax2.set_xlabel('X axis', fontsize = 15)
ax2.set_ylabel('Y axis', fontsize = 15)
ax2.set_title('Plot of:  polinomical function  ', fontsize = 13)
ax2.legend(loc = 'upper left', title = 'Info of functions')
ax2.set_xlim(-6, 6)
ax2.set_ylim( min(x2**2 + 2*x2 + 4), max(x2**2 + 2*x2 + 4) )
ax2.grid()

plt.tight_layout(pad = 1, h_pad = 1, w_pad = 1)
plt.show()
```

```python
x = np.arange(11)
y = x ** 2

fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = ( 14,8 ) )

axes[0,0].plot(x, y)
axes[0,0].set_title('Plot')
axes[0,0].set_xlabel('X')
axes[0,0].set_ylabel('Y')

axes[0,1].scatter(x, y)
axes[0,1].set_title('scatter')
axes[0,1].set_xlabel('X')
axes[0,1].set_ylabel('Y')

axes[0,2].bar(x, y)
axes[0,2].set_title('bar')
axes[0,2].set_xlabel('X')
axes[0,2].set_ylabel('Y')

axes[1,0].loglog(x, y)
axes[1,0].set_title('loglog')
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')

axes[1,1].semilogx(x, y)
axes[1,1].set_title('semilogx')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')

axes[1,2].semilogy(x, y)
axes[1,2].set_title('semilogy')
axes[1,2].set_xlabel('X')
axes[1,2].set_ylabel('Y')

plt.tight_layout()
plt.show()
plt.savefig('plots/test_plot.pdf')
```

# 3D PLOTTING

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
plt.style.use('seaborn-v0_8-poster')
#plt.subplots?
#%matplotlib notebook
```

```python
# Create the figure object
fig = plt.figure(figsize = (8,8))

# Creat and connect 'ax' an axe of plt with 'fig'
ax = fig.add_subplot(111, projection = '3d')

# Set grid to ax
ax.grid()

# Creating the point to plot and relationate
t = np.arange(0, np.pi * 10, np.pi / 50)
x = np.cos(t)
y = np.sin(t)

ax.set_title('TRigonometric Function')
ax.set_xlabel('x axis', labelpad = 20)
ax.set_ylabel('y axis', labelpad = 10)
ax.set_zlabel('z axis', labelpad = 15)

ax.plot(x, y, t, 'bo')


plt.show()
```

```python
# Create the figure
fig = plt.figure(figsize = (10,10))

# Adding trhe ax, 111 means col = row = inex = 1
ax = fig.add_subplot( 111, projection = '3d' )

# Creating points
t = np.random.random(220)
x = np.random.random(220)
y = np.random.random(220)

colors = np.sqrt(x**2 + y**2 + t**2)

ax.scatter(x, y, t, c = colors, cmap = 'coolwarm', s = 50)
ax.set_title('Random points')
ax.set_xlabel('x-axis', labelpad = 24)
ax.set_ylabel('y-axis', labelpad = 24)
ax.set_zlabel('z-axis', labelpad = 24)
```

```python
# Creating the points
ini = -5
fin = 5
n_data = 100
x = np.arange( ini, fin, (fin - ini) / n_data  )
y = np.arange( ini, fin, (fin - ini) / n_data )
X, Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)


# PLotting
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot( 111, projection = '3d' )
surface = ax.plot_surface(X, Y, Z, cmap = 'cividis')
fig.colorbar(surface, shrink = 0.5, aspect = 8)
```

```python
for i in range(3):
    print(f"Ciclo externo: i = {i}")
    for j in range(3):
        if j == 1:
            print(f"  Continue en j = {j}")
            continue  # Salta esta iteraciÃ³n del ciclo interno
        print(f"  Ciclo interno: j = {j}")
    
```

```python
x = 3
lista = [1,2,3,4,5]
while x > 0:
    for x in lista:
        if x == x:
            print(f"break statement, value of x: {x}")
            break
        print(x) 
    x = x-1
print(f"value of x after algorithm: {x}")
```
